```python
P = lambda t: 10*np.exp(2*t)

def dP (Pk,h):
    Pk1= Pk + h*2*Pk
    return Pk1

tfin = 3
h = .1 #Al generar un h mas pequeño, se aproxima más a la solución por Euler.

t = np.arange(0, tfin+.1, .1)
Pk = np.arange (0, tfin+h,h)

plt.plot(t,P(t), 'b')

Pk[0] = 10

for k in np.arange(0, Pk.size-1):
    Pk[k+1] = dP(Pk[k],h)
    #print (Pk[k+1])

tk = np.arange (0, tfin+h, h)
plt.plot(tk, Pk,'g')
plt.title ('Exponencial Euler vs Númerico', fontsize = 14)
tk = np.arange(0, tfin+h,h)

```
