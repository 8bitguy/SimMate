```python
# Numero de tirados aleatoriamente

N= 1000
f= lambda x: np.log(x)
#f= lambda x: np.sin(x)
#f= lambda x: np.cos(x)
a, b = 0, np.pi
Altura = 1

x = np.random.uniform(a,b,N)
y = np.random.uniform(a,Altura,N)
X = np.arange(a,b, .01)
fx= f(x)

plt.plot(x[y>fx],y[y>fx], "go")
plt.plot(x[y<fx],y[y<fx], "ro")
plt.plot(X,f(X))

N_0= len(x[y<fx])
Base = b-a
yest = N_0/N*(Base*Altura)
print (yest)
```
