```python
import numpy as np
```

```python
#Ver opcioones de plot (color, lineas, puntos, etc)
#plt.plot?
```

```python
t =  np.linspace(0, 50, 100)
plt.figure(figsize = (7,4))
plt.plot(t, .5*np.cos(.5*t) + .1 * np.sin(.5*t), '-', lw = 3, ms = 4,
         label = '$x(t)$')
plt.plot(t, .5*.1*np.cos(.5*t) -.5* .5* np.sin(.5*t), 'm.-', lw = 2, ms = 5, 
        label = r'$\dot{x(t)}$')
plt.xlabel('$t$', fontsize = 20)
plt.show()
```

```python
A = 2
w = .8
B = 0.1
t =  np.linspace(0, 50, 100)
plt.figure(figsize = (7,4))
plt.plot(t, A*np.cos(w*t) + B * np.sin(w*t), '-', lw = 3, ms = 4,
         label = '$x(t)$')
plt.plot(t, .5*.1*np.cos(.5*t) -.5* .5* np.sin(.5*t), 'm.-', lw = 2, ms = 5, 
        label = r'$\dot{x(t)}$')
plt.xlabel('$t$', fontsize = 20)
#plt.plot?
plt.legend(loc = 3)
plt.show()
```

```python
A = 2
B = 0.1
frecuencias = np.array([.1, .2, .5, .6])
#asignamos las amplitudes como f y hacemos un loop con el array frecuencias
plt.figure(figsize = (7,4))
for f in frecuencias:
    plt.plot(t, A * np.cos(f * t) + B * np.sin(f * t), '.-')
plt.xlabel('$t$', fontsize = 16)
plt.xlabel('$x(t)$', fontsize = 16)
plt.title('Oscilaciones', fontsize = 16)
plt.show()
```
