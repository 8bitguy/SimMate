## Desafío: caminante hacia arriba

- 10% de prob a la derecha
- 10% de prob a la izquierda
- 60% de prob a la arriba
- 20% de prob a la abajo

```python
def walker():
    global x,y
      
    eleccion = random.uniform(0,1)
    if eleccion < 0.10:
        x+=1
    elif eleccion < 0.10:
        x-=1
    elif eleccion < 0.60:
        y+=1
    else:
        y-=1
    return [x,y]
    
x = 0 #inicia el caminador
y = 0

pos = np.zeros((2,1000))

for  i in np.arange(0, len(pos[0]), 1):
    pos[:,i] = walker()

plt.plot(pos[0,:], pos[1,:], 'bo-')
plt.plot(pos[0,-1], pos[1, -1], 'ro-')
```

## Vuelo de Lévy

```python
def walker():
    global x,y
      
    r = random.uniform(0,1)
    if r < 0.01: #un 1% prob de tomar un paso grande
        x+=np.random.normal(0,200)
        y+=random.gauss(0,200)
    else:
        x+=np.random.normal(0,1)
        y+=random.gauss(0,1)
    return [x,y]
    
x = 0 #inicia el caminador
y = 0

pos = np.zeros((2,1000))

for  i in np.arange(0, len(pos[0]), 1):
    pos[:,i] = walker()

plt.plot(pos[0,:], pos[1,:], 'bo-')
plt.plot(pos[0,-1], pos[1, -1], 'ro-')
```
