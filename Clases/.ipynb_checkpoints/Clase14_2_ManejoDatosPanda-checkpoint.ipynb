```python
file_apple = 'Precios/AAPL.csv'
df_apple = pd.read_csv(file_apple, index_col='Date', usecols=('Date', 'Adj Close'))

file_amazon = 'Precios/AMZN.csv'
df_amazon = pd.read_csv(file_amazon, index_col='Date', usecols=('Date', 'Adj Close'))

file_nvidia = 'Precios/NVDA.csv'
df_nvidia = pd.read_csv(file_nvidia, index_col='Date', usecols=('Date', 'Adj Close'))

file_microsoft = 'Precios/MSFT.csv'
df_microsoft = pd.read_csv(file_microsoft, index_col='Date', usecols=('Date', 'Adj Close'))
```

```python
closes = pd.DataFrame(index=df_amazon.index, columns=['AAPL', 'AMZN', 'MSFT', 'NVDA'])
closes.index.name = 'Date'
closes['AAPL'] = df_apple
closes['AMZN'] = df_amazon
closes['MSFT'] = df_microsoft
closes['NVDA'] = df_nvidia
closes
```

```python
closes.plot(figsize =(10,6));
```

```python
#DEFINIMOS LOS INSTRUMENTOS QUE VAN A DESCARGAR DIRECTAMENTE DE YAHOO FINANCE
tickers = ['AAPL', 'AMZN', 'MSFT', 'NVDA']

#DEFINIMOS QUE FUENTE ONLINE VAMOS A USAR
data_source = 'yahoo'

#QUEREMOS LOS DATOS DESDE 01/01/2011 HASTA 31/12/2016.
start_date = datetime.datetime(2011,1,1)
end_date = datetime.datetime(2016,12,31)

#USAMOS LA FUNCION DATAREADER. SI, ASร DE FรCIL...
panel_data =  data.DataReader(tickers, data_source, start_date, end_date)
```

```python
panel_data
```

```python
closes = panel_data.loc['Adj Close']
closes
```

```python
closes.plot(figsize =(10,6));
```
